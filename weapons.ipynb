```python
# data preprocessing - 2nd step. Replacing categorical data to numerical using ordinal encoding

from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(dtype="int")

data.Type = encoder.fit_transform(data.Type.values.reshape(-1, 1))
data.Model = encoder.fit_transform(data.Model.values.reshape(-1, 1))
data.Condition = encoder.fit_transform(data.Condition.values.reshape(-1, 1))
data.Country = encoder.fit_transform(data.Country.values.reshape(-1, 1))
data.Buyer = encoder.fit_transform(data.Buyer.values.reshape(-1, 1))
```

```python
data
```

```python
# data splitting
from sklearn.model_selection import train_test_split
x = data.drop("Price(mln$)", axis=1)
y = data["Price(mln$)"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2, random_state=42, shuffle=True)
```

```python
# lets try Random Forest method to get a prediction

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
# params for RandomForestRegressor

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 150, 200],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 4],
    'max_features': ['sqrt', 'log2'],
    'criterion': ['squared_error', 'friedman_mse', 'poisson']
    }

grid_search = GridSearchCV(estimator=RandomForestRegressor(),
                           param_grid=param_grid, 
                           verbose=10, cv=3)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_
print(best_model.score(x_train, y_train))
```

```python
print(best_model.score(x_test, y_test))
```
